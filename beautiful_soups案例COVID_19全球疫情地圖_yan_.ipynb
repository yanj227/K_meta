```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://covid-19.nchc.org.tw/"
res = requests.get(url) 
soup = BeautifulSoup(res.text,'html.parser')#parser=html解析器
print(soup)
```

```python
total = soup.find(class_='country_confirmed mb-1 text-success')
clean_total = total.text.replace(',','')
print('累計確診:'+str(clean_total))
```


## 1 練習--擷取新增確診人數

###可以用ctrl+F找class標籤裡的字串
- 找到後貼到class_後面
- 會用class_是因為class是python內建的，為了不衝突才叫class_

```python
total_1 = soup.find(class_="country_confirmed mb-1 text-success")
clean_total_1 = total_1.text.replace(',','').replace('+','')

print('今日確診:'+str(clean_total_1))
```

```python
total_2 = soup.find(class_="country_deaths mb-1 text-dark")
clean_total_2 = total_2.text.replace(',','')

print('累計死亡:'+str(clean_total_2))
```

```python
for i in soup.find_all(class_='btn btn-success btn-lg'):
  link = 'https://covid-19.nchc.org.tw/'
  link = 'https://covid-19.nchc.org.tw/'+str(i.get('href'))
  #做成超連結，前面加上http://......
  print(link)
```

```python
import urllib#用於處理網址
```

##三列表分開處理


```python
#城市名稱列表
city_name = []
for i in soup.find_all(class_='btn btn-success btn-lg'):
  clean_link = i.get('href').split('=')
  city_name.append(clean_link[1])
print(city_name)
```

```python
#各城市網址列表
link_list = []
for j in range(len(city_name)):
  link = 'https://covid-19.nchc.org.tw/city_confirmed.php?mycity='+urllib.parse.quote(city_name[j])
  link_list.append(link)
print(link_list)
```

```python
#各城市新增確診人數法1(split('+'))
city_new_list = []
for i in soup.find_all(class_='btn btn-success btn-lg'):
  city_n=i.text.split('+')[1].strip()
  city_new_list.append(city_n)
print((city_new_list))
```

## 4 請將台灣本土病例分佈存成csv檔(其中檔案含二個欄位 )
- 縣市
- 連結網址


```python
import pandas as pd
df = pd.DataFrame()
df['縣市']=city_name
df['連結']=link_list
df.to_csv('/content/drive/MyDrive/Colab Notebooks/data/sample_data/cov19_data.csv', encoding='utf-8-sig', index=False)
```
